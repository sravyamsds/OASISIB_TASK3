```python
def create_histplot(df):
  for column in df.columns:
     median_value=np.median(df[column])
     print(f"{median_value} is median value for {column}.")
     sns.histplot(data=df,x=df[column]);
     plt.show()

```

```python
create_histplot(df)
```
